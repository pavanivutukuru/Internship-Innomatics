# Find number with Even Number of Digits

```python
class Digits:
    def __init__(self,numbers):
        self.numbers = numbers

    def Even1(self):
        output = []
        for num in self.numbers:
            if len(str(num)) % 2 == 0:
                output.append(num)
        return output
```

```python
list1 = [124,22,8654,796,5,36]           

D1 = Digits(list1)
result = D1.Even1()

print(result)
```

# Number of Good Pairs

```python
class Solution:
    def GoodPairs(self, nums):
        count = 0
        n = len(nums)

        for i in range(n):
            for j in range(i + 1, n):
                if nums[i] == nums[j]:
                    count += 1

        return count
```

```python
nums = [1,2,3,1,1,3]
obj = Solution()
print(obj.GoodPairs(nums))
```

# Subtract the Product and Sum of Digits of an Integer

```python
class Integer:
    def product(self,num):
        product = 1
        while num>0:
            X = num % 10
            product = product * X
            num = num // 10
            
        return product

    def sum1(self,num):
        total_sum = 0
        while num>0:
            Y = num % 10
            total_sum = total_sum + Y
            num = num//10
            
        return total_sum

    def final(self,num):
        p = self.product(num)
        s = self.sum1(num)

        print("Product of digits:", p)
        print("Sum of digits:", s)
        
        final = p - s
        print("The final value is :", final)
        return final
        
            
```

```python
N = Integer()
output = N.final(234)
```

```python
class Integer:
    def ProductandSum(self,num):
        product = 1
        total_sum = 0
        while num > 0:
            X = num % 10
            product = product * X
            total_sum = total_sum + X
            num = num // 10
            
        result = product - total_sum
        return result

N = Integer()
output = N.ProductandSum(234)
print("The final value is:", output)
```
