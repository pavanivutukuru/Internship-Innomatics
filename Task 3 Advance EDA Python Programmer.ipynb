# Count Number of Teams

```python
rating = [2, 5, 3, 4, 1, 6]
n = len(rating)

teams = [] 

for i in range(n):
    for j in range(i + 1, n):
        for k in range(j + 1, n):
            if (rating[i] < rating[j] < rating[k]) or (rating[i] > rating[j] > rating[k]):
                teams.append((rating[i], rating[j], rating[k]))

print("Total number of teams:", len(teams))
print("Teams list:")
for team in teams:
    print(team)
```

```python
class Teams:
    def __init__(self, rating):
        self.rating = rating
        self.n = len(rating)

    def find_teams(self):
        teams = []

        for i in range(self.n):
            for j in range(i + 1, self.n):
                for k in range(j + 1, self.n):
                    if (self.rating[i] < self.rating[j] < self.rating[k] or self.rating[i] > self.rating[j] > self.rating[k]):
                        teams.append((self.rating[i], self.rating[j], self.rating[k]))

        print("Total number of teams:", len(teams))
        print("Teams list:")
        return teams

```

```python
rating = [2, 5, 3, 4, 1, 6]

T1 = Teams(rating)
T1.find_teams()
```

# Number of Students Doing Homework at a Given Time

```python
startTime = [1, 2, 3]
endTime = [3, 2, 7]
queryTime = 4

count = 0

for i in range(len(startTime)):
    if startTime[i] <= queryTime <= endTime[i]:
        count += 1

print("Number of students doing homework:", count)
```

```python
class Homework:
    def __init__(self,startTime,endTime):
        self.startTime = startTime
        self.endTime = endTime

    def count1(self, querytime):
        count = 0

        for i in range(len(self.startTime)):
            if self.startTime[i] <= querytime <= self.endTime[i]:
                count += 1
        return count
```

```python
startTime = [1, 2, 3]
endTime = [3, 2, 7]
querytime = 4

obj = Homework(startTime, endTime)
result = obj.count1(querytime)

print("Number of students doing homework:", result)
```

# Number of Steps to Reduce a Number to Zero

```python
num = 14
steps = 0

while num > 0:
    if num % 2 == 0:
        num = num // 2
    else:
        num = num - 1
    steps += 1

print("Number of steps:", steps)
```

```python
 class Number:
     def Steps(self, num):
         steps = 0
         while num > 0:
             if num % 2 == 0:
                 num = num // 2
             else:
                 num = num - 1
             steps += 1
         return steps
```

```python
N = Number()
N.Steps(14)
```

# Counting Bits

```python
n = 5
ans = []

for i in range(n + 1):
    count = 0
    num = i
    while num > 0:
        count += num % 2
        num //= 2
    ans.append(count)

print(ans)
```

```python
class CountingBits:
    def count1(self, n):
        ans = [0] * (n + 1)

        for i in range(1, n + 1):
            ans[i] = ans[i // 2] + (i % 2)

        return ans
```

```python
C1 = CountingBits()
C1.count1(2)
```
