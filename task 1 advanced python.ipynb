 # Running sum of an array

```python
class Solution:
    def runningsum(self, nums):
        total = 0
        op = []
        for i in nums:
            total += i
            op.append(total)
        return op

```

```python
X = Solution()
X.runningsum([1,2,3,4,5])
```

# Shuffle the Array

```python
class Shuffle:
    def shuffle(self, nums, n):
        op = []
        for i in range(n):
            op.append(nums[i])
            op.append(nums[i + n])
        return op
```

```python
S1 = Shuffle()
S1.shuffle([2,5,1,3,4,7], 3)
```

# Kids with greatest number of candies

```python
class KidsWithCandies:
    def kidswithcandies(self, candies, extracandies):
        res = []
        max_val = max(candies)
        for i in candies:
            val = i+extracandies
            if max_val > val:
                res.append(False)
            else:
                res.append(True)
        return res
```

```python
K1 = KidsWithCandies()
K1.kidswithcandies([2,3,5,1,3],3)
```
